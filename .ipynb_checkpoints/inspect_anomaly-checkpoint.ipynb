```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
anomaly_df = pickle.load(open("anomaly_df.pickle", "rb"))
```

```python
anomaly_df.head()
```

```python
def open_with_gedit(anomaly_df, location):
    code_source = anomaly_df.loc[location].code_piece.src
    src_split_1 =  code_source.split(" : ")
    file_path = src_split_1[0]
    line_index = src_split_1[1].split(" - ")[0]
    bash_command = "gedit {} +{}".format(file_path, line_index)
    os.system(bash_command)
```

```python
anomaly_df.sort_values(by="anomaly_score", ascending=False)
```

```python
open_with_gedit(anomaly_df, 37)
```
